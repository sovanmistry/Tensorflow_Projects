## **Convert Numpy to Pytorch Tensors**

```python
# Creating Numpy array

import numpy as np

list = [1, 2, 3, 4]
arr = np.array(list)

arr
```

```python
arr.dtype
```

```python
# converting Numpy arr to pytorch tensor

tensor = torch.from_numpy(arr)
tensor
```

```python
tensor.dtype
```

```python
# Indexing - same as numpy

tensor[:2]
```

```python
tensor[2]
```

```python
tensor[1:4]
```

**Disadvantage from Numpy : both uses the same memory Loacation**

```python
tensor
```

```python
tensor[2] = 100
tensor
```

```python
arr
```

# **torch.tensor - Prevent the above problem**

```python
list = [1, 2, 3, 4]
arr = np.array(list)
```

```python
tensor_arr = torch.tensor(arr)
tensor_arr
```

```python
tensor_arr[2] = 100
tensor_arr
```

```python
arr
```

## **Mathematical Operation**

```python
a = torch.tensor([1, 2, 3, 4], dtype = torch.float)
b = torch.tensor([5, 6, 7, 8], dtype = torch.float)
```

```python
# Sum

torch.add(a,b)
```

```python
# Store the output


c = torch.zeros(1,4)
c
```

```python
torch.add(a,b, out = c)
```

```python
c
```

```python
# Sum of all elements - sum()

torch.add(a,b).sum()
```

```python
c.sum()
```

## **Matrix Multiplication**

```python
a = torch.tensor([[1, 2, 3, 4],[1, 2, 3, 4]])
b = torch.tensor([[5, 6],[7, 8],[9,10],[11,12]])
```

```python
# method 1

torch.matmul(a,b)
```

```python
# method 2

torch.mm(a,b)
```

```python
# method 3

a@b
```
