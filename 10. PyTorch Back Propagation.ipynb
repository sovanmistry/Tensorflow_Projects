## **PYTORCH BACK PROPAGATION**

```python
import torch
```

```python
x = torch.tensor(4.0, requires_grad = True)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
## BackPropagation y = 2*x

y.backward()
```

```python
print(x.grad)
```

```python
lst = [[1.,2.,3.],[1.,2.,3.],[1.,2.,3.]]
torch_input = torch.tensor(lst, requires_grad=True)
torch_input
```

```python
# y = x**3 + x**2

y = torch_input**3 + torch_input**2
```

```python
y
```

```python
z = y.sum()
```

```python
z.backward()
```

```python
torch_input.grad
```
